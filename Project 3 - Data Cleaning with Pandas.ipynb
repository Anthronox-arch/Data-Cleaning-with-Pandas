```python
# Percentage of missing values in each column.

for col in df_netflix.columns:
    percentage = df_netflix[col].isnull().mean()
    print(col + " -> " + str(round((percentage) * 100, 2)))
```

```python
df_netflix.drop('director', axis = 1) # "axis = 1" means column.
```

```python
# index_no_director = df_netflix[df_netflix['director'].isnull()].index
# df_netflix.drop(index_no_director, axis = 0) # "axis = 0" means rows (indexes)
```

```python
df_netflix[~(df_netflix['director'].isnull())] # Values that AREN'T NULL.
```

```python
df_netflix.dropna(subset = ['director'])
```

```python
mode = ''.join(df_netflix['rating'].mode()) # Find mode of 'rating' column. Mode = most repeating value.

# Rating column itself is a series, but the values inside are strings; so .join( ) can work correctly.
# We use mode since we have categorical and text data.

df_netflix['rating'] = df_netflix['rating'].fillna(mode)
```

```python
df_netflix['duration'] = df_netflix['duration'].fillna('0') # Arbitrary value '0'.
```

```python
# df_netflix.fillna(method = 'bfill') # Replace NaN with next value in column.
df_netflix.fillna(method = 'Ffill') # Replace NaN with previous value in the column.
```

```python
df_netflix_movie = df_netflix[df_netflix['type'] == 'Movie']
df_netflix_movie['minutes'] = df_netflix_movie['duration'].str.split(expand = True)[0].astype(int)

# Got an error here saying "Cannot convert from NaN to integer" because in the cell above, I used ".fillna(0)" (previously). So when I call .str,
# it replaces the integer 0 with NaN.
```

```python
# df_netflix_movie['date_added'].str.split(',', expand = True)[1] # Default separator is blank space.
df_netflix_movie['date_added'].str.extract('(\d{4})') # Regular expression. Also, requires capture group.
```

```python
# Identify outliers with histograms.

df_netflix_movie['minutes'].plot(kind = 'hist', xlabel = 'Length (Minutes)', title = "Movies' Length (Histogram/Frequency)", bins = 10)
```
